```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
from sklearn import datasets
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal_length','sepal_width','petal_length','petal_width','target'])
df
```

```python
columnname = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
x = df.loc[:, columnname].values
y = df.loc[:,['target']].values
```

```python
x
```

```python
y
```

```python
pca = PCA(n_components=3)
Parentvalue = pca.fit_transform(x)
parentdf = pd.DataFrame(data = Parentvalue)
Parentvalue
```

```python
parentdf
```

```python
resultDf = pd.concat([parentdf, df[['target']]], axis = 1)
resultDf
```

```python
xvals = resultDf[0]
yvals = resultDf[1]
zvals = resultDf[2]
```

```python
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel(0, fontsize = 16)
ax.set_ylabel(1, fontsize = 16)
ax.set_zlabel(2, fontsize = 16)
ax.set_title('iris data into 3 dimensions, ploted in 3d chart', fontsize = 24)

targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = resultDf['target'] == target
    ax.scatter(resultDf.loc[indicesToKeep, 0]
               , resultDf.loc[indicesToKeep, 1]
               , resultDf.loc[indicesToKeep, 2]
               , c = color
               , s = 100)
ax.legend(targets)
ax.grid()
plt.show()
```
